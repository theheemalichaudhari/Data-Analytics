# Importing libraries


```python
import pandas as pd
data = pd.read_csv("../input/masterclass/Walmart.csv")
data
```

# dropna removes the null cells
# The changes would be temoparary (in a way to check protect any mishappens )
# Inplace helps us to make the changes permanent by making the changes (trues makes the changes, by default it will be false)

```python
data.dropna(how="all",inplace=True)
data
```

# Here it fetches the country column from the data 
# Here to find the get the unique values we will use .unique()

```python
data['Country'].unique()
```

```python
data["Country"].replace({"United States":"USA"},inplace=True)
data
```

# Data Imputation
# Finding mean

```python
avg=data["Sales"].mean()
avg
```

```python
data["Sales"].fillna(value=avg,inplace=True)
data
```

# Exporting the file:
# Adding extension is important.

```python
data.to_csv("WalmartData.csv")
```
